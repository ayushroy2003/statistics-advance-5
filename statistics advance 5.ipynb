Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

Ans- 

```python
import scipy.stats as stats
import math

mean = 50
std_dev = 5
sample_size = 30 # Assumed

# z value for 95% confidendence interval 
z_value = 1.96

standard_error = std_dev / math.sqrt(sample_size)

lower_bound = mean-z_value*standard_error
upper_bound = mean+z_value*standard_error

print(f"for 95% Cofidence interval We Confident that the population mean lies between {lower_bound} ,{upper_bound}")
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

Ans - 

```python
observed_counts = [25,30,35,10,15,35]
expected_prop = [0.20,0.20,0.20,0.10,0.10,0.20]

total_m_ms = sum(observed_counts)

expected_counts = [total_m_ms * p for p in expected_prop]

chi_square_stat , p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

dof = len(observed_counts)-1

critical_value = stats.chi2.ppf(0.95,dof)

print("Chi-square statistics",chi_square_stat)
print("critical value", critical_value)
print("p value" ,p_value)

if chi_square_stat > critical_value:
    print("observed distribution is significantly different the expected distribution")
else:
    print("The observed distribution matches the expected distribution.")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

Ans- 

```python
import math
n = 500
smokers = 60
p_hat =  smokers / n

z = 1.96 # accordiing to z score table for 95% confidence ionterval 

se = math.sqrt((p_hat * (1- p_hat)) /n)

lower_bound = p_hat - z*se
upper_bound = p_hat + z*se

print("For 95% confidence interval : " , lower_bound, upper_bound)
print("This means that we are 95% confident that the true proportion of smokers in the population is between 9.15% and 14.85%")
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

Ans- 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10
chi_stat =15

x = np.linspace(0,30 ,500)
y = chi2.pdf(x, df)

plt.figure(figsize=(8,6))
plt.plot(x,y, label='Chi-Square Distribution (df=10)')

plt.fill_between(x, 0, y, where=(x <= chi_stat), color='lightblue', alpha=0.5)


plt.title('Chi-Square Distribution with 10 degree of freedom')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.axvline(chi_stat, color='red' , linestyle='--' , label=f'Chi-Square = {chi_stat}')

plt.legend()
plt.show()
```

Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis


A study was conducted to determine if there is an association between smoking status and lung cancer diagnosis. The data is given in the table below:

|                   | Lung Cancer: Yes | Lung Cancer: No | Row Total |
|-------------------|------------------|-----------------|-----------|
| **Smoker**        | 60               | 140             | 200       |
| **Non-smoker**    | 30               | 170             | 200       |
| **Column Total**  | 90               | 310             | 400       |

We will conduct a chi-square test for independence with a significance level of 0.05.

---

## Step 1: State the Hypotheses

- **Null Hypothesis (\\(H_0\\))**: There is no association between smoking status and lung cancer diagnosis.
- **Alternative Hypothesis (\\(H_1\\))**: There is an association between smoking status and lung cancer diagnosis.

---

## Step 2: Observed Frequencies

The observed frequencies are:

|                   | Lung Cancer: Yes | Lung Cancer: No | Row Total |
|-------------------|------------------|-----------------|-----------|
| **Smoker**        | 60               | 140             | 200       |
| **Non-smoker**    | 30               | 170             | 200       |
| **Column Total**  | 90               | 310             | 400       |

---

## Step 3: Calculate Expected Frequencies

The formula for expected frequencies is:

$$
E = \\frac{{(\\text{{Row Total}}) \\times (\\text{{Column Total}})}}{{\\text{{Grand Total}}}}
$$

Now, we calculate the expected frequencies:

- \\(E_{\\text{{Smoker, Yes}}} = \\frac{{200 \\times 90}}{{400}} = 45\\)
- \\(E_{\\text{{Smoker, No}}} = \\frac{{200 \\times 310}}{{400}} = 155\\)
- \\(E_{\\text{{Non-smoker, Yes}}} = \\frac{{200 \\times 90}}{{400}} = 45\\)
- \\(E_{\\text{{Non-smoker, No}}} = \\frac{{200 \\times 310}}{{400}} = 155\\)

The expected frequency table is:

|                   | Lung Cancer: Yes | Lung Cancer: No | Row Total |
|-------------------|------------------|-----------------|-----------|
| **Smoker**        | 45               | 155             | 200       |
| **Non-smoker**    | 45               | 155             | 200       |
| **Column Total**  | 90               | 310             | 400       |

---

### Step 4: Calculate the Chi-Square Test Statistic

The formula for the chi-square test statistic is:

$$
\\chi^2 = \\sum \\frac{{(O_i - E_i)^2}}{{E_i}}
$$

Where \\(O_i\\) are the observed frequencies and \\(E_i\\) are the expected frequencies.

Now, we calculate the chi-square value for each cell:

$$
\\chi^2_{\\text{{Smoker, Yes}}} = \\frac{{(60 - 45)^2}}{{45}} = \\frac{{15^2}}{{45}} = 5
$$
$$
\\chi^2_{\\text{{Smoker, No}}} = \\frac{{(140 - 155)^2}}{{155}} = \\frac{{(-15)^2}}{{155}} = 1.4516
$$
$$
\\chi^2_{\\text{{Non-smoker, Yes}}} = \\frac{{(30 - 45)^2}}{{45}} = \\frac{{(-15)^2}}{{45}} = 5
$$
$$
\\chi^2_{\\text{{Non-smoker, No}}} = \\frac{{(170 - 155)^2}}{{155}} = \\frac{{15^2}}{{155}} = 1.4516
$$

Summing these values:

$$
\\chi^2 = 5 + 1.4516 + 5 + 1.4516 = 12.9032
$$

---

### Step 5: Determine Degrees of Freedom

The degrees of freedom for a chi-square test for independence is calculated as:

$$
df = (r - 1)(c - 1)
$$

Where \\(r\\) is the number of rows and \\(c\\) is the number of columns.

Here:

$$
df = (2 - 1)(2 - 1) = 1
$$

---

### Step 6: Critical Value

From the chi-square distribution table, the critical value for a significance level of 0.05 and 1 degree of freedom is:

$$
\\chi^2_{\\text{{critical}}} = 3.841
$$

---

### Step 7: Make a Decision

We compare the test statistic with the critical value:

$$
\\chi^2_{\\text{{statistic}}} = 12.9032 > \\chi^2_{\\text{{critical}}} = 3.841
$$

Since the test statistic is greater than the critical value, we **reject the null hypothesis**.

---

### Conclusion

At the 0.05 significance level, we reject the null hypothesis. Therefore, there is sufficient evidence to conclude that there is an association between smoking status and lung cancer diagnosis.


Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark
chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from
the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the
contingency table below. Conduct a chi-square test for independence to determine if there is a significant
association between chocolate preference and country of origin.


A study was conducted to determine if the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K. The data is given in the table below:

|                   | Milk Chocolate | Dark Chocolate | White Chocolate | Row Total |
|-------------------|----------------|----------------|-----------------|-----------|
| **U.S.** (n=500)  | 200            | 150            | 150             | 500       |
| **U.K.** (n=500)  | 225            | 175            | 100             | 500       |
| **Column Total**  | 425            | 325            | 250             | 1000      |

We will conduct a chi-square test for independence with a significance level of 0.01.

---

### Step 1: State the Hypotheses

- **Null Hypothesis (\\(H_0\\))**: There is no association between chocolate preference and country of origin.
- **Alternative Hypothesis (\\(H_1\\))**: There is an association between chocolate preference and country of origin.

---

### Step 2: Observed Frequencies

The observed frequencies are:

|                   | Milk Chocolate | Dark Chocolate | White Chocolate | Row Total |
|-------------------|----------------|----------------|-----------------|-----------|
| **U.S.**          | 200            | 150            | 150             | 500       |
| **U.K.**          | 225            | 175            | 100             | 500       |

---

### Step 3: Calculate Expected Frequencies

The formula for expected frequencies is:

$$
E = \\frac{{(\\text{{Row Total}}) \\times (\\text{{Column Total}})}}{{\\text{{Grand Total}}}}
$$

Now, we calculate the expected frequencies:

- \\(E_{\\text{{U.S., Milk}}} = \\frac{{500 \\times 425}}{{1000}} = 212.5\\)
- \\(E_{\\text{{U.S., Dark}}} = \\frac{{500 \\times 325}}{{1000}} = 162.5\\)
- \\(E_{\\text{{U.S., White}}} = \\frac{{500 \\times 250}}{{1000}} = 125\\)
- \\(E_{\\text{{U.K., Milk}}} = \\frac{{500 \\times 425}}{{1000}} = 212.5\\)
- \\(E_{\\text{{U.K., Dark}}} = \\frac{{500 \\times 325}}{{1000}} = 162.5\\)
- \\(E_{\\text{{U.K., White}}} = \\frac{{500 \\times 250}}{{1000}} = 125\\)

The expected frequency table is:

|                   | Milk Chocolate | Dark Chocolate | White Chocolate | Row Total |
|-------------------|----------------|----------------|-----------------|-----------|
| **U.S.**          | 212.5          | 162.5          | 125             | 500       |
| **U.K.**          | 212.5          | 162.5          | 125             | 500       |
| **Column Total**  | 425            | 325            | 250             | 1000      |

---

### Step 4: Calculate the Chi-Square Test Statistic

The formula for the chi-square test statistic is:

$$
\\chi^2 = \\sum \\frac{{(O_i - E_i)^2}}{{E_i}}
$$

Where \\(O_i\\) are the observed frequencies and \\(E_i\\) are the expected frequencies.

Now, we calculate the chi-square value for each cell:

$$
\\chi^2_{\\text{{U.S., Milk}}} = \\frac{{(200 - 212.5)^2}}{{212.5}} = \\frac{{(-12.5)^2}}{{212.5}} = 0.7353
$$
$$
\\chi^2_{\\text{{U.S., Dark}}} = \\frac{{(150 - 162.5)^2}}{{162.5}} = \\frac{{(-12.5)^2}}{{162.5}} = 0.9615
$$
$$
\\chi^2_{\\text{{U.S., White}}} = \\frac{{(150 - 125)^2}}{{125}} = \\frac{{25^2}}{{125}} = 5
$$
$$
\\chi^2_{\\text{{U.K., Milk}}} = \\frac{{(225 - 212.5)^2}}{{212.5}} = \\frac{{12.5^2}}{{212.5}} = 0.7353
$$
$$
\\chi^2_{\\text{{U.K., Dark}}} = \\frac{{(175 - 162.5)^2}}{{162.5}} = \\frac{{12.5^2}}{{162.5}} = 0.9615

$$

Summing these values:

$$

The degrees of freedom for a chi-square test for independence is calculated as:

$$
df = (r - 1)(c - 1)
$$

Where \\(r\\) is the number of rows and \\(c\\) is the number of columns.

Here:

$$

### Step 6: Critical Value

From the chi-square distribution table, the critical value for a significance level of 0.01 and 2 degrees of freedom is:

$$

---

### Step 7: Make a Decision

We compare the test statistic with the critical value:

$$
\\chi^2_{\\text{{statistic}}} = 13.3936 > \\chi^2_{\\text{{critical}}} = 9.21
$$

Since the test statistic is greater than the critical value, we **reject the null hypothesis**.

---

### Conclusion

At the 0.01 significance level, we reject the null hypothesis. Therefore, there is sufficient evidence to conclude that there is an association between chocolate preference and country of origin.


Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

Ans - 
### Hypothesis Test for Population Mean

#### Given:
- Sample size, \(n = 30\)
- Sample mean, \(\bar{x} = 72\)
- Sample standard deviation, \(s = 10\)
- Hypothesized population mean, \(\mu_0 = 70\)
- Significance level, \(\alpha = 0.05\)

#### Steps:

1. **Set up the hypotheses:**

   - Null hypothesis (\(H_0\)): \(\mu = 70\) (The population mean is 70)
   - Alternative hypothesis (\(H_a\)): \(\mu \neq 70\) (The population mean is significantly different from 70)

2. **Test statistic formula:**

   $$
   t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}
   $$

   Substituting the values:

   $$
   t = \frac{72 - 70}{\frac{10}{\sqrt{30}}} = \frac{2}{\frac{10}{5.477}} = \frac{2}{1.826} \approx 1.095
   $$

3. **Degrees of freedom:**

   $$
   df = n - 1 = 30 - 1 = 29
   $$

4. **Critical value:**

   For a two-tailed test with \(\alpha = 0.05\) and \(df = 29\), the critical value is:

   $$
   t_{0.025, 29} \approx \pm 2.045
   $$

5. **Decision:**

   - Our calculated \(t\)-value is \(1.095\), which is less than the critical value \(2.045\).
   - Therefore, we **fail to reject the null hypothesis**.

6. **Conclusion:**

   At a significance level of 0.05, there is **not enough evidence** to conclude that the population mean is significantly different from 70.
